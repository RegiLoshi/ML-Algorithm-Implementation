```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming merged_data is your DataFrame with recipe features and ratings

features_to_analyze = ['n_steps', 'protein', 'calories', 'total_fat']

# 1. Quick correlation analysis
correlations = merged_data[features_to_analyze + ['rating']].corr()['rating'].sort_values(ascending=False)
print("Feature correlations with rating:")
print(correlations)

# 2. Simplified visualization
plt.figure(figsize=(12, 10))
for i, feature in enumerate(features_to_analyze, 1):
    plt.subplot(2, 2, i)
    sns.scatterplot(x=feature, y='rating', data=merged_data, alpha=0.1)
    plt.title(f'{feature} vs Rating')
plt.tight_layout()
plt.show()

# 3. Binned analysis
for feature in features_to_analyze:
    merged_data[f'{feature}_bin'] = pd.qcut(merged_data[feature], q=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
    avg_ratings = merged_data.groupby(f'{feature}_bin')['rating'].mean().sort_values(ascending=False)
    print(f"\nAverage ratings by {feature} bins:")
    print(avg_ratings)
```
